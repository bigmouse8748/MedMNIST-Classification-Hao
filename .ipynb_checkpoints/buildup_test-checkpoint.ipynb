# preprocess.py

```python
from medmnist import INFO, PathMNIST
from torchvision import transforms
from torch.utils.data import DataLoader
```

```python
def get_data_loaders(batch_size=32):
    # Define transforms
    train_transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(20),
        transforms.Normalize(mean=[.5], std=[.5])
    ])
    test_transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[.5], std=[.5])
    ])
    
    # Load datasets
    train_dataset = PathMNIST(split='train', transform=train_transform, download=True)
    test_dataset = PathMNIST(split='test', transform=test_transform, download=True)
    
    # Create DataLoaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    
    return train_loader, test_loader
```

## Train

```python
import torch
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from data.preprocess import get_data_loaders
from models.resnet_model import get_resnet
from models.inception_resnet import InceptionResNet
```

```python
def train_model(model_name, num_classes=9, epochs=10, batch_size=32, lr=0.001):
    # Select model
    if model_name == "resnet":
        model = get_resnet(num_classes)
    elif model_name == "inception_resnet":
        model = InceptionResNet(num_classes)
    else:
        raise ValueError("Invalid model name. Choose 'resnet' or 'inception_resnet'.")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)

    # Load data
    train_loader, test_loader = get_data_loaders(batch_size)

    # Define loss and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    # Training loop
    model.train()
    for epoch in range(epochs):
        running_loss = 0.0
        for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            # print(outputs)
            labels = labels.to(torch.long)
            # print(labels.squeeze())
            loss = criterion(outputs,labels.squeeze())
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        
        print(f"Epoch {epoch+1} Loss: {running_loss/len(train_loader)}")

    # Save the model
    torch.save(model.state_dict(), f"outputs/{model_name}_model.pth")
    print("Model saved!")

    return model
```

```python
import os
import time
import torch
import torch.nn as nn
import pandas as pd
from tqdm import tqdm
from data.preprocess import get_train_data_loaders
from models.stock_model import get_stock_model
from models.inception_resnet import InceptionResNet

MODEL_LIST = ["resnet50", "resnet18", "densenet121", "efficientnet_b0", "mobilenet_v3_large", "vgg16"]

def train_model(model_name, lr_strategy="default", pretrained=True, scheduler_type=None, num_classes=9, epochs=30, batch_size=32):
    # Timestamp for the training folder
    timestamp = time.strftime("%Y%m%d-%H%M%S")
    run_folder = os.path.join("outputs", f"{model_name}_{timestamp}")
    os.makedirs(run_folder, exist_ok=True)
    
    # Save training parameters to a CSV before training starts
    params = {
        "model_name": model_name,
        "lr_strategy": lr_strategy,
        "pretrained": pretrained if model_name in MODEL_LIST else "False",
        "scheduler_type": scheduler_type if scheduler_type else "None",
        "batch_size": batch_size,
        "epochs": epochs
    }
    params_df = pd.DataFrame([params])
    params_file = os.path.join(run_folder, "training_parameters.csv")
    params_df.to_csv(params_file, index=False)
    print(f"Training parameters saved to {params_file}")

    # Select model
    if model_name in MODEL_LIST:
        model = get_stock_model(model_name, num_classes, pretrained)
    elif model_name == "inception_resnet":
        model = InceptionResNet(num_classes)
    else:
        raise ValueError(f"Unsupported model_name: {model_name}")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)

    # Load data
    train_loader, val_loader = get_train_data_loaders()

    # Define learning rate strategies
    lr_options = {
        "default": 0.001,
        "low": 0.0001,
        "high": 0.01
    }
    if lr_strategy not in lr_options:
        raise ValueError(f"Invalid learning rate strategy. Choose from {list(lr_options.keys())}.")
    lr = lr_options[lr_strategy]

    # Define loss and optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    # Define learning rate scheduler
    scheduler = None
    if scheduler_type == "exponential":
        scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.9)
    elif scheduler_type == "ratio_decrease":
        def lr_lambda(epoch):
            return 0.5 if epoch > 5 else 1.0
        scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda=lr_lambda)

    # Logging setup
    log_file = os.path.join(run_folder, "training_log.csv")
    logs = []

    # Save the best model
    best_val_loss = float("inf")
    best_model_path = os.path.join(run_folder, f"best_model.pth")

    # Training loop
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
            inputs, labels = inputs.to(device), labels.to(device)
            labels = labels.squeeze()  # Ensure labels are 1D
            optimizer.zero_grad()
            outputs = model(inputs)
            labels = labels.to(torch.long)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        train_loss = running_loss / len(train_loader)

        # Validation loop
        model.eval()
        val_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                labels = labels.to(torch.long).squeeze()
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        val_loss /= len(val_loader)
        val_accuracy = correct / total

        # Save the best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), best_model_path)
            print(f"Best model saved at epoch {epoch+1} with validation loss {best_val_loss:.4f}")

        # Log training and validation results
        logs.append({
            "epoch": epoch + 1,
            "train_loss": train_loss,
            "val_loss": val_loss,
            "val_accuracy": val_accuracy,
        })
        print(f"Epoch {epoch+1}: Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}")

        # Step the scheduler
        if scheduler:
            scheduler.step()

        print(f"Learning Rate: {scheduler.get_last_lr()[0] if scheduler else lr}")

    # Save logs to CSV
    log_df = pd.DataFrame(logs)
    log_df.to_csv(log_file, index=False)
    print(f"Training log saved to {log_file}")

    return model
```

```python
model = train_model("mobilenet_v3_large", epochs = 5)
```
